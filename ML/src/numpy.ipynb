# Pre-Processing Data
In this section, we will discuss the steps involved in pre-processing data for machine learning models. Pre-processing is a crucial step in the machine learning pipeline as it helps to clean and prepare the data for analysis. The following are some common pre-processing techniques:
1. **Data Cleaning**: This involves handling missing values, removing duplicates, and correcting errors in the dataset. Techniques such as imputation can be used to fill in missing values, while duplicates can be removed to ensure the integrity of the data.
2. **Data Transformation**: This includes scaling, normalization, and encoding categorical variables. Scaling and normalization help to standardize the range of features, while encoding is used to convert categorical variables into a format that can be provided to machine learning algorithms.
3. **Feature Selection**: This process involves selecting the most relevant features for the model. This can be done using techniques such as correlation analysis, mutual information, or using algorithms like Recursive Feature Elimination (RFE).
4. **Data Splitting**: This involves dividing the dataset into training and testing sets. This is important to evaluate the performance of the machine learning model on unseen data. A common split is 80% for training and 20% for testing.


## Machine learning Models need NpArray or DataFrame ?

Machine learning models can work with both NumPy arrays and Pandas DataFrames, but the choice between the two depends on the specific requirements of the model and the data being used.
- **NumPy Arrays**: Many machine learning libraries, such as scikit-learn, primarily use NumPy arrays for their operations. NumPy arrays are efficient for numerical computations and can be easily manipulated for various machine learning tasks. If your data is purely numerical and does not require the additional functionality provided by Pandas, using NumPy arrays can be a good choice.
- **Pandas DataFrames**: Pandas DataFrames are more versatile and provide additional functionality for handling and manipulating data, especially when dealing with mixed data types (numerical and categorical). If your dataset contains categorical variables, missing values, or requires complex data manipulation, using a Pandas DataFrame can be more convenient. Additionally, many machine learning libraries can accept DataFrames directly, making it easier to work with the data without needing to convert it to a NumPy array.

### 3. Missing Values

Why do we need to handle missing values?
- Missing values can lead to biased estimates and reduce the statistical power of the model.
- Many machine learning algorithms cannot handle missing values and will throw an error if they encounter them.
- Handling missing values can improve the performance of the model and lead to better predictions.

## Sykit-learn provides several techniques for handling missing values, such as:
1. **Imputation**: This technique involves filling in the missing values with a specific value, such as the mean, median, or mode of the feature. The `SimpleImputer` class from `sklearn.impute` can be used for this purpose.

````python
### Importing Imputer
from sklearn.impute import SimpleImputer
### Creating Object and specifying what we are going to replace with what Strategy
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
### NOTE: This need to be performed againts the numerical value only
### Column 2 and 3 are numerical so added to fit
### Calculating with FIT
imputer.fit(X[:,1:3])
### Replacing with TRANSFORM
X[:,1:3] = imputer.transform(X[:,1:3])
````

Using Fit and Transform method of SimpleImputer class, we can handle missing values in our dataset. The `fit` method is used to calculate the imputation values (e.g., mean, median) based on the training data, while the `transform` method is used to apply the imputation to the dataset. Alternatively, we can use the `fit_transform` method to perform both steps in one go. This allows us to effectively handle missing values and prepare our data for machine learning models.

Possible Strategies for Imputation:
- **Mean Imputation**: Replacing missing values with the mean of the feature.
- **Median Imputation**: Replacing missing values with the median of the feature.
- **Mode Imputation**: Replacing missing values with the mode of the feature (most frequent value).

Selecting the appropriate imputation strategy depends on the nature of the data and the distribution of the feature. For example, mean imputation is suitable for normally distributed data, while median imputation is more robust to outliers. Mode imputation is typically used for categorical features.

Selecting Columns with Missing Values:
To identify columns with missing values in a dataset, we can use the `isnull()` function in combination with `sum()` to count the number of missing values in each column. This allows us to focus our imputation efforts on the relevant features and ensure that we are addressing the missing data effectively.

### 4. Data Transformation
Data transformation is a crucial step in the machine learning pipeline that involves modifying the data to make it more suitable for analysis. This can include techniques such as scaling, normalization, and encoding categorical variables. Scaling and normalization help to standardize the range of features, while encoding is used to convert categorical variables into a format that can be provided to machine learning algorithms.

#### 4.1 One-Hot Encoding
One-hot encoding is a technique used to convert categorical variables into a format that can be provided to machine learning algorithms. It creates new binary columns for each category in the original variable, where a value of 1 indicates the presence of that category and a value of 0 indicates its absence. This allows machine learning models to effectively utilize categorical data without assuming any ordinal relationship between the categories.

For Example if we have a categorical variable "Color" with three categories: Red, Green, and Blue, one-hot encoding would create three new binary columns: "Color_Red", "Color_Green", and "Color_Blue". If a data point has the value "Red" for the "Color" variable, the corresponding row in the new columns would be [1, 0, 0], indicating that the color is Red and not Green or Blue.



````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
columTransformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[0])], remainder='passthrough')
X = np.array(columTransformer.fit_transform(X))
````

````python
print(X)
````

#### 4.2 Label Encoding
Label encoding is another technique used to convert categorical variables into a format that can be provided to machine learning algorithms. It assigns a unique integer to each category in the original variable. This can be useful for ordinal categorical variables where there is a natural order between the categories. However, it can introduce unintended relationships between categories if used on nominal categorical variables, as the model may interpret the integer values as having a meaningful order.

This can be done using the `LabelEncoder` class from `sklearn.preprocessing`. For example, if we have a categorical variable "Color" with three categories: Red, Green, and Blue, label encoding would assign the values 0, 1, and 2 to these categories respectively. If a data point has the value "Red" for the "Color" variable, it would be encoded as 0, while "Green" would be encoded as 1 and "Blue" as 2.

Best use of this technique is when the categorical variable has an inherent order, such as "Low", "Medium", and "High". In this case, label encoding would assign the values 0, 1, and 2 to these categories respectively, which reflects their natural order. However, for nominal categorical variables like "Color", one-hot encoding is generally preferred to avoid introducing unintended relationships between categories.

#### Do we need to apply Feature Scaling before split or after split?


Feature scaling should be applied after splitting the dataset into training and testing sets. This is because we want to ensure that the scaling is based only on the training data, and not influenced by the testing data. If we apply feature scaling before splitting, we may inadvertently introduce information from the testing set into the training process, which can lead to overfitting and biased performance evaluation. By applying feature scaling after splitting, we can ensure that our model is trained on appropriately scaled data while maintaining the integrity of the testing set for unbiased evaluation.

### 6. Feature Scaling
Feature scaling is a technique used to standardize the range of features in a dataset. This is important because many machine learning algorithms are sensitive to the scale of the features, and having features on different scales can lead to suboptimal performance. Common feature scaling techniques include normalization (min-max scaling) and standardization (z-score scaling). Normalization scales the features to a range of [0, 1], while standardization centers the features around the mean with a standard deviation of 1. The choice of scaling technique depends on the specific requirements of the machine learning algorithm being used and the distribution of the data.

Type of Scaling:
- **Normalization (Min-Max Scaling)**: This technique scales the features to a range of [0, 1]. It is calculated using the formula: (X - min) / (max - min). Normalization is useful when the data does not follow a Gaussian distribution and when we want to preserve the shape of the original distribution.

<b>Value lies between 0 and 1 after normalization.</b>

- **Standardization (Z-Score Scaling)**: This technique centers the features around the mean with a standard deviation of 1. It is calculated using the formula: (X - mean) / standard_deviation. Standardization is useful when the data follows a Gaussian distribution and when we want to maintain the shape of the original distribution.

<b>Value lies between -3 and 3 after standardization.</b>

What is Gaussian Distribution?
A Gaussian distribution, also known as a normal distribution, is a continuous probability distribution that is symmetric around its mean. It is characterized by its bell-shaped curve, where the mean, median, and mode are all equal. The Gaussian distribution is defined by two parameters: the mean (μ) and the standard deviation (σ). The mean determines the center of the distribution, while the standard deviation controls the spread of the data. Many natural phenomena and measurement errors tend to follow a Gaussian distribution, making it a fundamental concept in statistics and machine learning.

<b>Do we apply feature scalling on Dummy Variables?</b>

No, we do not apply feature scaling on dummy variables (one-hot encoded variables) because they are already in a binary format (0 and 1) and do not require scaling. Scaling dummy variables can distort their meaning and lead to incorrect interpretations by the machine learning model. Since dummy variables represent categorical information, they should be left as is to preserve their interpretability and ensure that the model can effectively utilize them for predictions.
