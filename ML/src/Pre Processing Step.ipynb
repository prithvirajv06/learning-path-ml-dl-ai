# Pre-Processing Data
In this section, we will discuss the steps involved in pre-processing data for machine learning models. Pre-processing is a crucial step in the machine learning pipeline as it helps to clean and prepare the data for analysis. The following are some common pre-processing techniques:
1. **Data Cleaning**: This involves handling missing values, removing duplicates, and correcting errors in the dataset. Techniques such as imputation can be used to fill in missing values, while duplicates can be removed to ensure the integrity of the data.
2. **Data Transformation**: This includes scaling, normalization, and encoding categorical variables. Scaling and normalization help to standardize the range of features, while encoding is used to convert categorical variables into a format that can be provided to machine learning algorithms.
3. **Feature Selection**: This process involves selecting the most relevant features for the model. This can be done using techniques such as correlation analysis, mutual information, or using algorithms like Recursive Feature Elimination (RFE).
4. **Data Splitting**: This involves dividing the dataset into training and testing sets. This is important to evaluate the performance of the machine learning model on unseen data. A common split is 80% for training and 20% for testing.

### What is Feature and Depended Variable
In machine learning, a **feature** (also known as an independent variable) is an individual measurable property or characteristic of a phenomenon being observed. Features are the input variables that are used to make predictions or classifications in a machine learning model. For example, in a dataset of house prices, features could include the size of the house, the number of bedrooms, and the location.


A **dependent variable** (also known as a target variable) is the outcome or the variable that we are trying to predict or explain. It is dependent on the features. In the house price example, the dependent variable would be the price of the house, which we are trying to predict based on the features.

### 3. Missing Values

Why do we need to handle missing values?
- Missing values can lead to biased estimates and reduce the statistical power of the model.
- Many machine learning algorithms cannot handle missing values and will throw an error if they encounter them.
- Handling missing values can improve the performance of the model and lead to better predictions.

## Sykit-learn provides several techniques for handling missing values, such as:
1. **Imputation**: This technique involves filling in the missing values with a specific value, such as the mean, median, or mode of the feature. The `SimpleImputer` class from `sklearn.impute` can be used for this purpose.

```python
### Importing Imputer
from sklearn.impute import SimpleImputer
### Creating Object and specifying what we are going to replace with what Strategy
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```
